```python
def foo(x, y):
    return evaluate([x, 120, 7, 0, 1, 0, y], layers=2)
```

```python
%matplotlib notebook
limitsdict = {'r_limits':(10,120), 'm_limits':(3, 10), 'mj_limits':(0, 0.4)}
r1 = (10, 120)
r2 = (10, 120)

xs = np.linspace(r1[0], r1[1], 100)
ys = np.linspace(r2[0], r2[1], 100)
Xs, Ys = np.meshgrid(xs, ys)
Zs = np.vectorize(foo)(Xs, Ys)

fig = plt.figure()#figsize=(16,10))
ax = plt.axes(projection='3d')
#ax.contour3D(Xs, Ys, Zs, 50, cmap='binary')
ax.plot_surface(Xs, Ys, Zs, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
